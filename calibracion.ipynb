```python
import cv2
import matplotlib.pyplot as plt
```

```python
rutas = ['CALIBRACION/IMG_0017.JPG','CALIBRACION/IMG_0018.JPG','CALIBRACION/IMG_0019.JPG',
            'CALIBRACION/IMG_0020.JPG','CALIBRACION/IMG_0021.JPG','CALIBRACION/IMG_0022.JPG',
            'CALIBRACION/IMG_0023.JPG','CALIBRACION/IMG_0024.JPG','CALIBRACION/IMG_0025.JPG']
```

```python
imagenes = [cv2.imread(ruta) for ruta in rutas]
plt.figure(figsize=(10, 10))
for i, imagen in enumerate(imagenes,start=1):
    plt.subplot(3,3,i)
    plt.imshow(imagen[:,:,::-1])


```

```python
for i in range(len(imagenes)):
    imagen_gris = cv2.cvtColor(imagenes[i], cv2.COLOR_BGR2GRAY)
    retval, corners = cv2.findChessboardCorners(imagen_gris, (4,6), None) #detectar esquinas de la imagen en escala de grises 
    retval
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(imagen_gris, cmap='gray')
plt.show()
```
